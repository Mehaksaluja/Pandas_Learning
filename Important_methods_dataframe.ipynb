```python
# Find the last match played by virat kohli in delhi
ipl.head()
```

```python
ipl['all_players'] = ipl['Team1Players'] + ipl['Team2Players']
ipl.head(2)
```

```python
def virat_kohli_played(players):
  return 'V Kohli' in players
```

```python
ipl['virat_kohli_played'] = ipl['all_players'].apply(virat_kohli_played)
ipl[(ipl['City'] == 'Delhi') & (ipl['virat_kohli_played'] == True)].drop_duplicates(subset = ['City','virat_kohli_played'],keep='first')
```

```python
# apply in dataframes
points = pd.DataFrame({
    'First_point' : [(1,2),(3,4),(5,6),(7,8)],
    'Second_point' : [(1,5),(2,7),(5,6),(7,8)]
})

points
```

```python
def distance(row):
  pt_1 = row['First_point']
  pt_2 = row['Second_point']
  return ((pt_2[0] - pt_1[0])**2 + (pt_2[1] - pt_1[1])**2)**0.5
```

```python
points.apply(distance,axis = 1)
```
